```python
import pandas as pd
import seaborn as sea
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.pyplot as plt

# Imports for Machine Learning 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import tensorflow as tf
```

```python
# Copy original data frame into a new one to clean and drop columns/rows with a large amount of null values
hotel2_df = hotels_df.copy()
hotel2_df.drop(columns=['agent','company'],inplace=True)
hotel2_df.isnull().sum()
```

```python
# Dropping null values in columns 'children' and 'country'
hotel2_df = hotel2_df.dropna()
hotel2_df.isnull().sum()
```

```python
hotel2_df.to_csv('hotel_2_data.csv',index=False)
```

```python
clean_hotel_df = hotel2_df.copy()
clean_hotel_df.drop(columns = ['reservation_status_date','reserved_room_type','assigned_room_type',
                           'arrival_date_week_number','distribution_channel','country'],inplace = True)
clean_hotel_df.head(10)
```

```python
clean_hotel_df.to_csv('clean_hotel_df.csv',index = False)
```

```python
clean_hotel_df.columns
```

```python
# Making a copy of the df
df_enc = clean_hotel_df.copy()
```

```python
# Encoding month column values
months_num = {
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12,
}


df_enc['arrival_date_month'] = df_enc['arrival_date_month'].apply(lambda x : months_num[x])
df_enc.head()
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
hotel_cat = df_enc.dtypes[df_enc.dtypes == 'object'].index.tolist()

df_enc[hotel_cat].nunique()
```

```python
# Create OneHotEncoder instance 
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(df_enc[hotel_cat]))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(hotel_cat)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
df_enc = df_enc.merge(encode_df,left_index=True, right_index=True)
df_enc = df_enc.drop(hotel_cat,1)
df_enc.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = df_enc["is_canceled"].values
X = df_enc.drop(["is_canceled"],1).values

from sklearn.model_selection import train_test_split
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

Model Building: Logistic Regression

```python
# Create our logistic regression model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs', max_iter=100, random_state=1)
classifier
```

```python
# Training the model
classifier.fit(X_train_scaled, y_train)
```

```python
# Create predictions
y_pred = classifier.predict(X_test_scaled)
pd.DataFrame({"Prediction": y_pred, "Actual": y_test})
```

```python
# Validate the model's performance
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
# Validating the model using the test data
print(f"Training Data Score: {classifier.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test_scaled, y_test)}")
```

```python
# Confusion Matrix 
from sklearn.metrics import confusion_matrix, classification_report
matrix = confusion_matrix(y_test, y_pred)
print(matrix)
```

```python
# Classification Report
report = classification_report(y_test, y_pred)
print(report)
```

Model Building: Random Forrest Classifier

```python

from sklearn.ensemble import RandomForestClassifier
```

```python
# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=128, random_state=1) 
```

```python
# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data.
predictions = rf_model.predict(X_test_scaled)
```

```python
# Calculating the confusion matrix.
cm = confusion_matrix(y_test, predictions)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])

cm_df
```

```python
# Calculating the accuracy score.
acc_score = accuracy_score(y_test, predictions)
acc_score
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```
