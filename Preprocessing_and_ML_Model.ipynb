```python
# Copy original data frame into a new one to clean and drop columns/rows with a large amount of null values
hotel2_df = hotels_df.copy()
hotel2_df.drop(columns=['agent','company'],inplace=True)
hotel2_df.isnull().sum()
```

```python
# Dropping null values in columns 'children' and 'country'
hotel2_df = hotel2_df.dropna()
hotel2_df.isnull().sum()
```

```python
#New version: keep booking changes
clean_hotel_df = hotel2_df.copy()
clean_hotel_df.drop(columns = ['reservation_status_date','reserved_room_type','assigned_room_type',
                           'arrival_date_week_number','distribution_channel','country','reservation_status',
                           'arrival_date_year','arrival_date_day_of_month'],inplace = True)
clean_hotel_df.head(10)
```

```python
clean_hotel_df.to_csv('clean_hotel_df.csv',index = False)
```

```python
clean_hotel_df.columns
```

```python
# Making a copy of the df
df_enc = clean_hotel_df.copy()
```

```python
# Encoding month column values
months_num = {
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12,
}


df_enc['arrival_date_month'] = df_enc['arrival_date_month'].apply(lambda x : months_num[x])
df_enc.head()
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
hotel_cat = df_enc.dtypes[df_enc.dtypes == 'object'].index.tolist()

df_enc[hotel_cat].nunique()
```

```python
# Create OneHotEncoder instance 
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(df_enc[hotel_cat]))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(hotel_cat)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
df_enc = df_enc.merge(encode_df,left_index=True, right_index=True)
df_enc = df_enc.drop(hotel_cat,1)
df_enc.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = df_enc["is_canceled"].values
X = df_enc.drop(["is_canceled"],1).values

from sklearn.model_selection import train_test_split
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

Model Building: Logistic Regression

```python
# Create our logistic regression model
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression(solver='lbfgs', max_iter=100, random_state=1)
lr_model
```

```python
# Training the model
lr_model.fit(X_train_scaled, y_train)
```

```python
# Create predictions
y_pred = lr_model.predict(X_test_scaled)
pd.DataFrame({"Prediction": y_pred, "Actual": y_test})
```

```python
# Validate the model's performance
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
# Validating the model using the test data
print(f"Training Data Score: {lr_model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {lr_model.score(X_test_scaled, y_test)}")
```

```python
# Confusion Matrix 
from sklearn.metrics import confusion_matrix, classification_report
matrix = confusion_matrix(y_test, y_pred)
print(matrix)
```

```python
# Classification Report
report = classification_report(y_test, y_pred)
print(report)
```

Model Building: Random Forrest Classifier

```python

from sklearn.ensemble import RandomForestClassifier
```

```python
# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=128, random_state=1) 
```

```python
# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data.
predictions = rf_model.predict(X_test_scaled)
```

```python
# Calculating the confusion matrix.
cm = confusion_matrix(y_test, predictions)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])

cm_df
```

```python
# Calculating the accuracy score.
acc_score = accuracy_score(y_test, predictions)
acc_score
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```

SMOTEENN

```python
# Use the SMOTEENN technique to perform combination sampling on the data
# Count the resampled classes
from imblearn.combine import SMOTEENN
from collections import Counter

smote_enn = SMOTEENN(random_state=0)
X_resampled, y_resampled = smote_enn.fit_resample(X_train_scaled, y_train)
Counter(y_resampled)
```

Logistic Regression

```python
# Fit a Logistic regression model using random undersampled data
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression(solver='lbfgs', random_state=1)
model_lr.fit(X_resampled, y_resampled)
```

Evaluation Metrics

```python
# Display the confusion matrix
from sklearn.metrics import confusion_matrix

y_pred = model_lr.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
# Calculate the Balanced Accuracy Score
from sklearn.metrics import balanced_accuracy_score

balanced_accuracy_score(y_test, y_pred)
```

```python
# Print the imbalanced classification report
from imblearn.metrics import classification_report_imbalanced

print(classification_report_imbalanced(y_test, y_pred))
```

## Testing Gradient Boosting Classifier to improve our recall percentage

Choosing best learning rate

```python
from sklearn.ensemble import GradientBoostingClassifier

# Create a classifier object
learning_rates = [0.05, 0.1, 0.25, 0.5, 0.75, 1]
for learning_rate in learning_rates:
    classifier = GradientBoostingClassifier(n_estimators=350,
                                            learning_rate=learning_rate,
                                            max_features=15,
                                            max_depth=10,
                                            random_state=0)

    # Fit the model
    classifier.fit(X_train_scaled, y_train)
    print("Learning rate: ", learning_rate)

    # Score the model
    print("Accuracy score (training): {0:.3f}".format(
        classifier.score(
            X_train_scaled,
            y_train)))
    print("Accuracy score (validation): {0:.3f}".format(
        classifier.score(
            X_test_scaled,
            y_test)))
    print()
```

Choose Learning rate

```python
# Choose a learning rate and create classifier
# In our case the learning rate with the highest accuracy score after validation is 1
gbc_model = GradientBoostingClassifier(n_estimators=350,
                                        learning_rate=0.25,
                                        max_features=15,
                                        max_depth=10,
                                        random_state=0)

# Fit the model
gbc_model.fit(X_train_scaled, y_train)

# Make Prediction
predictions = gbc_model.predict(X_test_scaled)
pd.DataFrame({"Prediction": predictions, "Actual": y_test}).head(10)
```

Evaluate the model

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
acc_score = accuracy_score(y_test, predictions)
print(f"Accuracy Score : {acc_score}")
```

```python
cm = confusion_matrix(y_test, predictions)
cm_df = pd.DataFrame(
   cm, index=["Actual 0", "Actual 1"],
   columns=["Predicted 0", "Predicted 1"]
)
display(cm_df)
```
