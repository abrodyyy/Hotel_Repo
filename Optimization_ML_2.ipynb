```python
# Import libraries
import pandas as pd
```

```python
# Reading in dataset
file_path = r"C:\Users\aleah\Class\Final_Project\Hotel_Repo\Resources\hotel_bookings.csv"
df = pd.read_csv(file_path)
df.head(10)
```

```python
# Copy of original df first
hotels_df_2 = df.copy()

# Dropping the 'agent' and 'company' columns 
hotels_df_2.drop(columns = ['agent', 'company'], inplace=True)
hotels_df_2.columns
```

```python
# Dropping the 4 ROWS with missing 'children' data
# Dropping the 488 ROWS with missing 'country' data
hotels_df_2 = hotels_df_2.dropna()

# Rechecking NaN values 
for column in hotels_df_2.columns:
    print(f"Column {column} has {hotels_df_2[column].isnull().sum()} null values.")
```

```python
# Dropping more columns that may not be necessary for our analysis 
# Danira dropped: 'reservation_status_date','reserved_room_type','assigned_room_type','arrival_date_week_number','distribution_channel','country'
# Keeping 'reserved_room_type', 'distribution_channel','country'
# Keeping "arrival_date_year", "arrival_date_day_of_month", "booking_changes" (we dropped it in the shared colab file)
hotels_df_2.drop(columns = ["reservation_status_date", "assigned_room_type", "arrival_date_week_number"], inplace=True)

# dropping reservation_status too
hotels_df_2.drop(columns = ["reservation_status"], inplace=True)
hotels_df_2.columns
```

```python
import sklearn as skl

# Prepping country column for encoding
country_counts = hotels_df_2.country.value_counts()
country_counts.head(15)
```

```python
# Determine which values to replace
replace_countries = list(country_counts[country_counts < 2000].index)

# Replace in DataFrame
for item in replace_countries:
    hotels_df_2.country = hotels_df_2.country.replace(item,"Other")


# Check to make sure binning was successful
hotels_df_2.country.value_counts()
```

```python
# Encoding month column values
months_num = {
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12,
}


hotels_df_2['arrival_date_month'] = hotels_df_2['arrival_date_month'].apply(lambda x : months_num[x])
hotels_df_2.head()
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder

hotel_cat = hotels_df_2.dtypes[hotels_df_2.dtypes == 'object'].index.tolist()

hotels_df_2[hotel_cat].nunique()
```

```python
# Create OneHotEncoder instance 
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(hotels_df_2[hotel_cat]))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(hotel_cat)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
hotels_df_2 = hotels_df_2.merge(encode_df,left_index=True, right_index=True)
hotels_df_2 = hotels_df_2.drop(hotel_cat,1)
hotels_df_2.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = hotels_df_2["is_canceled"].values
X = hotels_df_2.drop(["is_canceled"],1).values

from sklearn.model_selection import train_test_split
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create our logistic regression model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs', max_iter=100, random_state=1)
classifier
```

```python
# Training the model
classifier.fit(X_train_scaled, y_train)
```

```python
# Create predictions
y_pred = classifier.predict(X_test_scaled)
pd.DataFrame({"Prediction": y_pred, "Actual": y_test})
```

```python
# Validate the model's performance
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
# Validating the model using the test data
print(f"Training Data Score: {classifier.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test_scaled, y_test)}")
```

```python
# Random Forrest
from sklearn.ensemble import RandomForestClassifier

# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=128, random_state=78) 

# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data.
predictions = rf_model.predict(X_test_scaled)
```

```python
# Calculating the confusion matrix.
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(y_test, predictions)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])

cm_df
```

```python
# Calculating the accuracy score.
acc_score = accuracy_score(y_test, predictions)
acc_score
```
