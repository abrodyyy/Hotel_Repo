```python
# Importing libraries
import pandas as pd
```

```python
# Reading in dataset
file_path = r"C:\Users\aleah\Class\Final_Project\hotel_bookings.csv"
hotels_df = pd.read_csv(file_path)
hotels_df.head(10)
```

```python
# Exploring the columns 
hotels_df.columns
```

```python
# Splitting dataset into 2

# Resort hotel
resort_hotel_df = hotels_df.loc[hotels_df['hotel'] == 'Resort Hotel']
resort_hotel_df.describe()
```

```python
# City hotel
city_hotel_df = hotels_df.loc[hotels_df['hotel'] == 'City Hotel']
city_hotel_df.describe()
```

```python
# Number of cancellations at the Resort Hotel by month
resort_cancellations_month = resort_hotel_df.groupby(['arrival_date_month']).sum()['is_canceled']
resort_cancellations_month.sort_values(ascending=False)
```

```python
# Number of cancellations at the City Hotel by month
city_cancellations_month = city_hotel_df.groupby(['arrival_date_month']).sum()['is_canceled']
city_cancellations_month.sort_values(ascending=False)
```

```python
# Plotting cancellations 
hotels_df.plot(x="arrival_date_month", y=["resort_cancellations_month"], kind="bar")
```

```python
import matplotlib.pyplot as plt

resort_cancellations_month.sort_values(ascending=False).plot(kind='bar')
```

```python
city_cancellations_month.sort_values(ascending=False).plot(kind='pie', autopct='%.1f%%')
```

```python
# Mean ADR (average daily rate) of cancellations vs non-cancellations
# (SHOULD I BE MULTIPLYING THE ADR BY TOTAL NIGHTS STAYED/BOOKED?)
adr_per_canellations = hotels_df.loc[hotels_df['is_canceled'] == 1, 'adr'].mean()
adr_per_noncancellations = hotels_df.loc[hotels_df['is_canceled'] == 0, 'adr'].mean()

print(f"The mean average daily rate for canceled reservations was: {adr_per_canellations}")
print(f"The mean average daily rate for non-canceled reservations was: {adr_per_noncancellations}")
```
